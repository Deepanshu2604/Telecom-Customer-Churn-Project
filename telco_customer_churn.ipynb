## convert TotalCharges to numeric datatype

```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'] , errors='coerce')
```

## remove missing values

```python
df.dropna(inplace=True)
```

## convert churn to numeric

```python
df['Churn']=df['Churn'].map({'Yes':1,'No':0})
df.head()
```

## calculate churn rate

```python
churn_rate=df['Churn'].mean()*100                  # churn_rate=26% means 26 out of 100 left the company
print(churn_rate)
```

## Exploratory Data Analysis (EDA)

## 1. Churn by Contract

```python
sns.countplot(x='Contract',hue='Churn',data=df)
plt.title("Churn by Contract Tyoe")
plt.show()
```

```python
# Insight:  month-to-month customers churn the most
```

## 2. Churn vs Monthly Charges

```python
sns.boxplot(x='Churn' ,y='MonthlyCharges',data=df)
plt.show()
```

```python
# insight: high monthly charges ---> high churn risk
```

## 3. Churn vs Tenure

```python
sns.histplot(data=df,x='tenure',hue='Churn',bins=30)
plt.show()
```

```python
# Insights: New customer leaves more often
```
